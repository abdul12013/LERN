one-hot-encodin = is a type of string encoding use when the data are few or in a limited range like yes, no  and married, unmarried etc

```python
import pandas as pd
```

```python
data=pd.read_csv('/content/1_1_2023_general.csv')
data.head(20)
```

```python
data.isnull().sum()
```

```python
for i in data.select_dtypes(include='object').columns:
  print(i)
  data[i].fillna(data[i].mode()[0],inplace=True)
```

```python
data.isnull().sum()
```

```python
en_data=data[['type']]
```

One-hot -encoding get dumies technique

```python
pd.get_dummies(data=en_data)
```

one-hot-ecoding using scite-learn

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder(drop='first')
arr=ohe.fit_transform(en_data).toarray()
```

```python
dat=pd.DataFrame(arr,columns=ohe.get_feature_names_out())
```
