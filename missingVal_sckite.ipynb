Filling missing value using skite learn

```python
from sklearn.impute import SimpleImputer
```

```python
si=SimpleImputer(strategy='mean')
arr=si.fit_transform(data[['amount_amount', 'hourly_rate', 'client_payment_verification_status',
       'client_total_feedback', 'client_total_spent']])
```

```python
updated=pd.DataFrame(arr,columns=data.select_dtypes(include='float64').columns)
```
